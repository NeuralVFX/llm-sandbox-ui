# https://github.com/NeuralVFX/unreal-llm-sandbox

> Jupyter Notebook style IDE for Unreal, with LLM Prompting and Agents


## Features

- **Notebook Interface** - Markdown, Code, Prompt, and Agent cells
- **Unreal Engine Integration** - Execute Python directly in UE5.6 via remote kernel
- **LLM Tool Calling** - Agents can search the web, read URLs, and iterate on code
- **Auto-saving** - Notebooks save to `.ipynb` format every 2 seconds



### Installation

pip install unreal-llm-sandbox

Or from GitHub:
pip install git+https://github.com/NeuralVFX/unreal-llm-sandbox.git

## Quick Start

### Documentation

Launch the notebook server
`start_server()`

Then open `http://localhost:5001/notebook/my_notebook.ipynb`

## Cell Types

| Type | Purpose |
|------|---------|
| **Markdown** | Documentation and notes |
| **Code** | Python executed in Unreal Engine |
| **Prompt** | Single LLM query with context |
| **Agent** | Iterative code generation with unit testing |
